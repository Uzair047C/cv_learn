```python
import tensorflow as tf
import cv2
from ultralytics import YOLO

```

```python

model = YOLO("yolov8n.pt")

cap = cv2.VideoCapture(0)

while True:

    ret, frame = cap.read()

    results = model(frame)

    annotated = results[0].plot()

    cv2.imshow("Detection", annotated)

    if cv2.waitKey(1) == ord('q'):
        break
```
